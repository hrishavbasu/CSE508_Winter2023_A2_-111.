```python
def ndcg(df, n):
    # compute the Discounted Cumulative Gain (DCG) score
    dcg = 0
    for i in range(n):
        dcg += (pow(2, df[0][i]) - 1) / np.log2(i + 2)
    return dcg
```

```python
ndcg(my_df,50)/ndcg(final_df,50)
```

```python
ndcg(my_df,len(my_df))/ndcg(final_df,len(my_df))
     
```
